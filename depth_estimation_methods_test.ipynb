```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
depth_image= cv.imread('./res/depth_map_1732876096.277.png')
segmention_image= cv.imread('./seg/nnnseg_overlay.png')
```

```python
# visualize the depth image in rgb
depth_image = cv.cvtColor(depth_image, cv.COLOR_BGR2GRAY)
segmention_image = cv.resize(segmention_image, (depth_image.shape[1], depth_image.shape[0]), interpolation=cv.INTER_NEAREST)

plt.imshow(depth_image, cmap='gray')
plt.title('Depth Image')
print('Depth Image shape:', depth_image.shape)


```
